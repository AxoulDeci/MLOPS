```python
#Chargement des librairies
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Chargement de la base de données
AD = pd.read_csv('N_RECODAGE.csv')
```

```python
#Visualisation de la base de données
AD.head()
```

```python
#Pays d'origine des individus enquêtés
modalites_obs= AD['Pays_natal'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
modalites_obs.plot(kind='bar', color=couleurs.colors)
plt.title('Répartion du pays d\'origine')
plt.xlabel('modalités')
plt.ylabel('nombres d\'occurences')
plt.show() 
# Cette visualisation permet de voir la répartition du pays d'origine des individus enquêtés, en effet nous remarquons la quasi totalité sont originaires des états-unis
```

```python

#Revenu annuel des inidividus enqêtés
modalites_obs= AD['Income'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
plt.pie(modalites_obs, labels=modalites_obs.index, autopct='%1.1f%%', colors=couleurs.colors)
plt.title('Répartition des revenus')
plt.show()
#Cette visualisation permet de voir la répartition des revenus annuels des individus enquêtés, en effet nous remarquons que juste une minorité arrivent à percevoir un salaire annuel supérieur à 50 000$ 
```

```python

#Niveau scolaire des individus enquêtés
modalites_obs= AD['Education'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
modalites_obs.plot(kind='bar', color=couleurs.colors)
plt.title('Répartion du niveau scolaire')
plt.xlabel('modalités')
plt.ylabel('nombres d\'occurences')
plt.show() 
#Cette visualisation permet de voir la répartition du niveau scolaire des individus enquêtés, en effet nous remarquons une bonne repartition ce qui montre les individus ont généralement un niveau scolaire raisonnable
```

```python
#Le sexe des personnes enquêtées
modalites_obs= AD['Sexe'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
plt.pie(modalites_obs, labels=modalites_obs.index, autopct='%1.1f%%', colors=couleurs.colors)
plt.title('Répartitionn du Sexe')
plt.show()
#Cette visualisation permet de voir la répartition du sexe des individus enquêtés, en effet on fait face à une population mixte avec une majorité d'hommes
```

```python
#Le nombre d'heures de travail par semaine des personnes enquêtées
modalites_obs= AD['Heures_par_sem'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
plt.pie(modalites_obs, labels=modalites_obs.index, autopct='%1.1f%%', colors=couleurs.colors)
plt.title('Nombre d\'heures de travail par semaine')
plt.show()
#Cette visualisation peremt d'avoir du nombre d'heures/semaine éffectués généralement par les personnes enquêtées, effet une grosse partie travaillent généralment 40H ou moins 
```

```python
#La race des personnes enquêtées
modalites_obs= AD['Race'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
modalites_obs.plot(kind='bar', color=couleurs.colors)
plt.title('Race de la population')
plt.xlabel('modalités')
plt.ylabel('nombres d\'occurences')
plt.show()
#Cette visualisation permet d'avoir une idée de la race des personnes enquêtées, en effet on fait face à une population quasiment que de race Blanche ce qui nous fait croire que ces données ne sont pas récentes 
```

```python
#Le secteur d'activité des personnes enquêtées
modalites_obs= AD['Emploi_Secteur'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
plt.pie(modalites_obs, labels=modalites_obs.index, autopct='%1.1f%%', colors=couleurs.colors)
plt.title('Secteur d\'activité')
plt.show()
#Cette visualisation nous pouvons de voir les différents secteur d'activité dans lesquels travaillent les personnes enquêtées, en effet nous remarquons la majorité travaillent dans le secteur privé
```

```python
#Le type de métier des individus enquêtés
modalites_obs= AD['Occupation'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
modalites_obs.plot(kind='bar', color=couleurs.colors)
plt.title('Type de métier')
plt.xlabel('modalités')
plt.ylabel('Nombre d\'occurences')
plt.show() 
#Cette visualisation nous montre le type de métier exercé par les personnes enquêtées, en effet nous retrouvons 5 métiers qui ressortent de l'enquête avec une forte concentration au niveau des métiers de l'administration et de l'artisanat mécanique
```

```python
#La relation matrimoniale des personnes enquêtées
modalites_obs= AD['Relation_familiale'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
plt.pie(modalites_obs, labels=modalites_obs.index, autopct='%1.1f%%', colors=couleurs.colors)
plt.title('Relation matrimoniale')
plt.show()
#Cette visualisation nous montre la situation relationnelle des personnes enquêtées, en effet nous remarquons qu'une quasi moitié sont conjoints et une autre catégorie un peu plus que la moitié sont autres
```

```python
#Le statut matrimonial des personnes enquêtées
modalites_obs= AD['Statut_marital'].value_counts()
couleurs=plt.cm.get_cmap('tab10',len(modalites_obs))
plt.pie(modalites_obs, labels=modalites_obs.index, autopct='%1.1f%%', colors=couleurs.colors)
plt.title('Statut matrimonial')
plt.show()
#Cette visualisation nous montre le statut matrimonial des personnes enquêtées, en effet nous retouvons 3 modalités (Marie,Divorce,Seul) avec une majorité qui sont mariées
```

```python
#Croisement entre le Revenu annuel et le type de métier
cross_tab_percent = pd.crosstab(AD['Income'], AD['Occupation'])
print(cross_tab_percent)
#Nous constatons que la majorité des personnes touchant un salaire supérieur à 50 000$ travaillent dans les métiers administratifs
```

```python
#Croisement entre le Revenu annuel et le niveau scolaire
cross_tab = pd.crosstab(AD['Income'], AD['Education'])
print(cross_tab)
#Nous constatons que la majorité des personnes touchant un salaire supérieur sont allés au déla des études supérieurs comme par exemple le doctorat
```

```python
#Croisement entre le Revenu annuel et le secteur d'activité
cross_tab = pd.crosstab(AD['Income'], AD['Emploi_Secteur'])
print(cross_tab)
#Nous constatons que la majorité des personnes touchant un salaire supérieur à 50 000$ travaillent dans le secteur Privé
```

```python
#Croisement entre le Revenu annuel et le nombre d'heure/semaine
cross_tab = pd.crosstab(AD['Income'], AD['Heures_par_sem'])
print(cross_tab)
#On constate que la majorité des personnes touchant un salire supérieur à 50 000$ travaillent généralement 40H ou moins par semaine mais exercent forcément dans les meilleurs secteurs
```
